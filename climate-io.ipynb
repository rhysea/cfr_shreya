# Input/Output

In this tutorial, we introduce the input and output of climate datasets with `cfr`.

```python
%load_ext autoreload
%autoreload 2

import cfr
import pandas as pd
import numpy as np
import os

import warnings
warnings.filterwarnings('ignore')
```

## Load test data to `ClimateField`

```python
dirpath = 'data/GCM_sims/icesm_last_millennium_historical/'
fd_tas = cfr.ClimateField().load_nc(os.path.join(dirpath, 'tas_sfc_Amon_iCESM_past1000historical_085001-200512.nc'))
fd_tas.da
```

## Save `ClimateField` to a NetCDF file

```python
lat = np.arange(-90, 90.1, 10)
lon = np.arange(0, 360, 10)
fd_tas_lres = fd_tas.regrid(lat, lon)
fd_tas_lres.to_nc('./data/fd_tas_lres.nc')
```

```python
import xarray as xr

fd = xr.open_dataarray('./data/fd_tas_lres.nc')
fd
```
